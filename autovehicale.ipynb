```python
def plotpolly(model, independent_variable, depend_variable,Name):
    x_new = np.linspace(15,55,100)
    y_new = model(x_new)
    
    plt.plot(independent_variable,depend_variable,'.',x_new,y_new,'-')
    plt.title('Polynomial Fit with Matplotlib')
    ax = plt.gca()
    ax.set_facecolor((0.898,0.898,0.898))
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')
    
    plt.show()
    plt.close()
```

```python
def Distribution(RedFunction, BlueFunction, RedName,BlueName,Title):
 
    
    ax1 = sns.distplot(RedFunction,hist=False,color='r',label=RedName)
    ax2 = sns.distplot(BlueFunction,hist=False,color='b',label=BlueName,ax=ax1)
    
    plt.title(Title)
    plt.xlabel('Price (in dollars)')
    plt.ylabel('proporstion of cars')
    
    plt.show()
    plt.close()
```

```python
def PollyPlot(x_train,x_test,y_train,y_test,lr,ploy_transform):
    xmax = max([x_train.values.max(),x_test.values.max()])
    
    xmin = min([x_train.values.max(),x_test.values.max()])
    
    x = np.arange(xmin,xmax,0.1)
    
    plt.plot(x_train,y_train,'ro',label='Training Data')
    plt.plot(x_test,y_test,'go',label='Test Data')
    plt.plot(x,lm7.predict(poly_transform.fit_transform(x.reshape(-1,1))), label = 'predicted Function')
    plt.ylim([-10000,60000])
    plt.ylabel('Price')
    plt.legend()
    
```
