```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [2,4,6,8,10,12,14]

plt.figure(figsize=(7,7),dpi=100)

plt.plot(x,y,label='2X',color='g', marker='^',linewidth=2,markersize=13,markerfacecolor='black', linestyle='solid')



plt.xlabel('X Lable',fontdict={'fontname':'Comic Sans MS','fontsize':15,'color':'blue'})
plt.ylabel('Y Lable',fontdict={'fontname':'Comic Sans MS','fontsize':15,'color':'blue'})


plt.xticks([1,2,3,4,5,6,7,8])
plt.yticks([2,4,6,7,8,10,12,14,15])

x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:5],x2[:5]**2, label= 'X2**2 First Value',color='r',linestyle='-')

plt.plot(x2[4:],x2[4:]**2, label= 'X2**2 last Value',color='b',linestyle='--')

plt.legend()

plt.title('Training Graph', fontdict={'fontname': 'Comic Sans MS','fontsize':15,'color':'red'})

plt.savefig('mygraph.png',dpi=300)

plt.show()
```

### bar chart

```python
lables = ['A','B','C']
values = [2,6,4]

bars= plt.bar(lables,values)

plt.figure(figsize=(7,7),dpi=100)

bars[0].set_hatch('*')
bars[1].set_hatch('o')
bars[2].set_hatch('/')

plt.show()
```
