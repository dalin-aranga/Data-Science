```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas.head()
```

```python
plt.figure(figsize=(10,6),dpi=100)

plt.title('Gas Prices over Time (in USD)',fontdict={'fontsize':16,'fontweight':'bold'})

for i in gas.columns:
    if i !='Year':
        plt.plot(gas.Year,gas[i],label = 'South Korea',marker='o',linewidth=2,markersize=6, linestyle='solid')

plt.xlabel('Year',fontdict={'fontsize':14,'fontweight':'bold'})
plt.ylabel('Price (USD $)',fontdict={'fontsize':12,'fontweight':'bold'})

plt.xticks(gas.Year[::3].tolist()+[2011])


plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6),dpi=100)

plt.title('Gas Prices over Time (in USD)',fontdict={'fontsize':16,'fontweight':'bold'})

plt.plot(gas['Year'],gas['Canada'], label = 'Canada',marker='^',linewidth=2,markersize=12, linestyle='solid')
plt.plot(gas['Year'],gas['Australia'], label = 'Australia',marker='^',linewidth=2,markersize=12, linestyle='solid')
plt.plot(gas['Year'],gas['South Korea'], label = 'South Korea',marker='^',linewidth=2,markersize=12, linestyle='solid')


plt.xlabel('Year',fontdict={'fontsize':14,'fontweight':'bold'})
plt.ylabel('Price (USD $)',fontdict={'fontsize':12,'fontweight':'bold'})

plt.xticks(gas.Year[::3].tolist()+[2011])


plt.legend()
plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
fifa
```

```python
print(fifa.isnull().sum())
```

```python
fifa.columns
```

```python
bins = [30,40,50,60,70,80,90,100]

plt.figure(figsize=(10,6),dpi=100)

plt.hist(fifa.Overall, bins=bins)

plt.xticks(bins)

plt.ylabel('Number of players')
plt.xlabel('overall skill')
plt.title('Distribution of Plyer Skill in FIFA 1018',fontdict={'fontsize':12,'fontweight':'bold'})


plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

plt.figure(figsize=(10,6),dpi=100)

plt.pie([left,right],labels=['Left','Right'],colors=['g','y'],autopct='%.2f%%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight.head()
```

```python
light = fifa.loc[fifa.Weight<125].count()[0]
light_medium = fifa.loc[(fifa.Weight>=125)&(fifa.Weight<150)].count()[0]
medium = fifa.loc[(fifa.Weight>=150)&(fifa.Weight<175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight>=175)&(fifa.Weight<200)].count()[0]
heavy = fifa.loc[fifa.Weight>200].count()[0]

plt.figure(figsize=(10,6),dpi=100)

plt.style.use('ggplot')

weight = [light,light_medium,medium,medium_heavy,heavy]
lables = ['Uder 125','125-150','150-175','175-200','Over 200']

explode = (.6,.1,.0,.0,.5)

plt.pie(weight,labels=lables,autopct='%.2f%%',pctdistance=0.8,explode =explode )

plt.title('Weght Destribution Of FIFA Playes (in lbs)',fontdict={'fontname': 'Comic Sans MS','fontsize':15,'color':'black'})

plt.show()
```

```python
plt.style.use('default')

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

plt.figure(figsize=(6,7),dpi=100)

lables = ['FC Barcelona','Real Madrid','New England Revolution']

boxes = plt.boxplot([barcelona,madrid,revs],labels=lables,patch_artist=True)

for box in boxes['boxes']:
    box.set(color = 'b',linewidth = 2,facecolor = '#e0e0e0')
    

plt.title('Professional Score Team Comparison',fontdict={'fontname': 'Comic Sans MS','fontsize':15,'color':'black'})
plt.ylabel('FIFA Overall Rating')

plt.show()
```
