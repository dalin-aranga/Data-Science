```python
def count_plot(colums, fig):
    plt.figure(figsize=(15,20))
    plt.subplot(4,2,fig)
    sns.countplot(x=colums,data = cars)
    plt.title(f'{colums} Count Plot',color='black',fontdict={'fontsize':18,'fontweight':'bold'})

    plt.subplot(4,2,fig+1)
    sns.boxplot(x=cars[colums],y=cars.price)
    plt.title(f'{colums} vs Price Box Plot',color='black',fontdict={'fontsize':18,'fontweight':'bold'})
    plt.show()
```

```python
count_plot('enginelocation',1)
count_plot('cylindernumber',3)
count_plot('fuelsystem',5)
count_plot('drivewheel',7)
```

1. Very few datapoints for enginelocation categories to make an inference.
2. Most common number of cylinders are four, six and five. Though eight cylinders have the highest price range.
3. mpfi and 2bbl are most common type of fuel systems. mpfi and idi having the highest price range. But there are few data for other categories to derive any meaningful inference
4. A very significant difference in drivewheel category. Most high ranged cars seeme to prefer rwd drivewheel.

```python
def numerical_data(columns,y):
    plt.figure(figsize=(15,15))
    plt.subplot(2,2,y)
    sns.scatterplot(x=cars[columns], y=cars.price)
    plt.title(f'{columns} vs price  Scatter Plot')
    
```

```python
numerical_data('carlength',1)
numerical_data('carwidth',2)
numerical_data('carheight',3)
numerical_data('curbweight',4)
```

1. carwidth, carlength and curbweight seems to have a poitive correlation with price.
2. carheight doesn't show any significant trend with price.
